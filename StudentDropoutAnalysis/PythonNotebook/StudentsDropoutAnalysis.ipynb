```python
#FEATURE ENGINEERING
#PASS RATIO
df['PassRatio'] = df['TotalUnitsApproved'] / df['TotalUnitsEnrolled']
```

```python
#FINANCIAL STRESS
df['FinancialStress'] = ((df['Debtor'] == 1) | (df['TuitionFeesUpToDate'] == 0)).astype(int)
```

```python
df['GradeDiff_2ndMinus1st'] = df['CurricularUnits2ndtSem_grade'] - df['CurricularUnits1stSem_grade']
```

```python
# FOCUSED CORRELATION HEATMAP (Key Features)
plt.figure(figsize=(10,6))
sns.heatmap(df[['PassRatio','FinancialStress','CurricularUnits1stSem_grade','CurricularUnits2ndtSem_grade','AgeAtEnrollment']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# MODEL TRAINING AND PREDICTION
# FEATURE SELECTION
x = df[['PassRatio', 'FinancialStress', 'AgeAtEnrollment']]
y = df['Target']

# LABEL ENCODING
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# TRAIN-TEST SPLIT
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y_encoded, test_size=0.2, random_state=42
)

# MODEL TRAINING
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
model.fit(x_train, y_train)

print(f"Training set: {x_train.shape[0]} samples")
print(f"Test set: {x_test.shape[0]} samples")

# PREDICTION
y_pred = model.predict(x_test)

# MODEL EVALUATION

from sklearn.metrics import classification_report, confusion_matrix

# MODEL EVALUATION: CONFUSION MATRIX
print('\n Confusion Matrix:\n')
print(confusion_matrix(y_test, y_pred))
print('\nConfusion Matrix shows the model predicts Dropout and Graduate well, but struggles with Enrolled.')

# MODEL EVALUATION: CLASSIFICATION REPORT
print('\n Classification Report:\n')
print(classification_report(y_test, y_pred, target_names=le.classes_))
print('\n Classification Report indicates F1-score is low for Enrolled (0.04) but decent for Dropout (0.79) and Graduate (0.80).')

# MODEL INTERPRETATION: COEFFICIENTS
coeff_df = pd.DataFrame(model.coef_, columns=x.columns, index=le.classes_)
print('\n Model Coefficients: \n')
print(coeff_df)
print('\nCoefficients highlight PassRatio and FinancialStress as key drivers of student outcomes.')
```

```python
# VISUALIZATION: IMPACT OF FEATURES ON STUDENT OUTCOMES
coeff_df = pd.DataFrame(model.coef_, columns=x.columns, index=le.classes_)
coeff_df = coeff_df.T

# PLOT
plt.figure(figsize=(10, 6))
coeff_df.plot(kind='bar', figsize=(10,6), colormap='viridis')
plt.title('Impact of Features on Student Outcomes', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Coefficient Value (Impact on Outcome)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Outcome', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
Key Socioeconomic Insight (Business-Ready):

- PassRatio and FinancialStress are the most important factors affecting student success.

- AgeAtEnrollment has little impact in comparison.

- Schools should support students with low academic performance and provide financial assistance to reduce dropout risk.
```
